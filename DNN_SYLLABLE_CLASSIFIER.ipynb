# Apply DNN to MFCCs

```python
import pandas as pd
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense,Dropout
from sklearn.metrics import accuracy_score

df = pd.read_csv('/content/syllable_mfcc_trained.csv')
rf = pd.read_csv('/content/syllable_mfcc_untrained.csv')
```

```python
df=df.drop(['Unnamed: 0','39'],axis=1)
rf=rf.drop(['Unnamed: 0','39'],axis=1)
```

```python
label_tr = df['40']
label_te = rf['40']
df=df.drop(['40'],axis=1)
rf=rf.drop(['40'],axis=1)

label_tr
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np 
le = LabelEncoder()
label_te = le.fit_transform(label_te)
label_tr = le.fit_transform(label_tr)
np.unique(label_te)
```

```python
from keras.utils import to_categorical

label_te = to_categorical(label_te, num_classes=40)
label_tr = to_categorical(label_tr, num_classes=40)
```

```python
label_tr
```

```python
df
```

```python
model = Sequential()
model.add(Dense(200, input_dim=39, activation='relu'))
model.add(Dense(2500,activation='relu'))
model.add(Dropout(0.2, input_shape=(2500,)))
model.add(Dense(2500,activation='relu'))
model.add(Dense(40, activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
print(model.summary())
```

```python
model.fit(df,label_tr,epochs=20,batch_size=32)
```

```python
y_predict_mlp = []
for i in range(len(rf)):
  y_predict_mlp.append(model.predict(np.array([rf.iloc[i]])).argmax())
```

```python
# confusion matrix for the multi layer perceptron model 
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
from sklearn import metrics
unique_labels = np.unique(label_te)
mlp_cm=metrics.confusion_matrix(label_te,y_predict_mlp, labels=unique_labels)
disp = plot_confusion_matrix(mlp_cm,cmap=plt.cm.Blues)
plt.show()
print()
ppn_cm = pd.DataFrame(mlp_cm,columns=unique_labels,index= unique_labels)
ppn_cm.style.background_gradient(cmap ='viridis')
```
